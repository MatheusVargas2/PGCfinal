# Gráficos Threshold

### Cria gráficos threshold (mods 1, 2 e 3)

```python
def training(mod, k, print_results=False):
    tentativas_legitimo = 0
    sucesso_legitimo = 0
    tentativas_ataque_interno = 0
    ataques_interno = 0
    tentativas_ataque_externo = 0
    ataques_externo = 0

    for i in mod[5]:
        tentativas_legitimo += 1
        if i < k:
            sucesso_legitimo += 1
    
    for i in mod[6]:
        tentativas_ataque_interno += 1
        if i < k:
            ataques_interno += 1

    for i in mod[7]:
        tentativas_ataque_externo += 1
        if i < k:
            ataques_externo += 1
            
    FNMR = (tentativas_legitimo-sucesso_legitimo)/tentativas_legitimo
    FMR_Int = ataques_interno / tentativas_ataque_interno
    FMR_Ext = ataques_externo/tentativas_ataque_externo
    FMR = (ataques_interno+ataques_externo)/(tentativas_ataque_interno+tentativas_ataque_externo)
    BAcc = 1 - (FMR + FNMR)/2

    if print_results:
        print("k:            ", k)
        print("FNMR:          {0:.2%}".format(FNMR))
        print("FMR Interno:   {0:.2%}".format(FMR_Int))
        print("FMR Externo:   {0:.2%}".format(FMR_Ext))
        print("FMR:           {0:.2%}".format(FMR))
        print("BAcc:          {0:.2%}".format(BAcc))
    
    return [k, FNMR, FMR_Int, FMR_Ext, FMR, BAcc]

def plota_resultados(modelo_, inf, sup, qtd, img, k=0):
    k_list = []
    for i in np.linspace(inf, sup, qtd):
        k_list.append(training(modelo_, i))
    results = pd.DataFrame(k_list, columns=["k", "FNMR", "FMR_Int", "FMR_Ext", "FMR", "BAcc"])
    plt.plot(np.linspace(inf, sup, qtd), results["FNMR"], 'g')
    # plt.plot(np.linspace(inf, sup, qtd), results["BAcc"], 'b')
    plt.plot(np.linspace(inf, sup, qtd), results["FMR_Int"], 'r')
    plt.plot(np.linspace(inf, sup, qtd), results["FMR_Ext"], 'm')
    plt.legend(["FNMR", "FMR Internos", "FMR Externos"], loc='upper center')
    plt.ylim([0, 0.1])
    plt.gca().set_yticklabels(['{:.1f}%'.format(x*100) for x in plt.gca().get_yticks()]) 
    plt.xlabel("threshold k")
    plt.savefig("threshold_" + str(img))
    plt.show()

    if k > 0:
        a = training(modelo_, k)
        print("FNMR: {0:.2%}".format(a[1]))
        print("FMR: {0:.2%}".format(a[4]))
        print("FMR_Int: {0:.2%}".format(a[2]))
        print("FMR_Ext: {0:.2%}".format(a[3]))
        print("BAcc: {0:.2%}".format(a[5]))
```
